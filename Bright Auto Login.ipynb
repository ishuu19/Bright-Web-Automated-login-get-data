### Selenium Imports
#### ChromeDriver and Chorme version - 131

```python
import json
import time
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
```

## Code

### Only Login

```python
file_path = 'accounts.txt'

accounts = []
with open(file_path, 'r') as file:
    lines = file.readlines()
    for i in range(1, len(lines)):  
        splitList = lines[i].strip().split(",") 
        if len(splitList) == 2: 
            accounts.append({"username": splitList[0], "password": splitList[1]})


options = Options()
options.add_argument("--start-maximized") 
service = Service("C:/Users/User/Documents/chromedriver-win64/chromedriver-win64/chromedriver.exe") 

driver = webdriver.Chrome(service=service, options=options)

login_url = "https://bright.glowmarkt.com/login"


for account in accounts:
    try:
        driver.get(login_url)
        WebDriverWait(driver, 7).until(
            EC.visibility_of_element_located((By.NAME, "ion-input-0"))
        )
        
        username_input = driver.find_element(By.NAME, "ion-input-0")
        username_input.clear()
        username_input.send_keys(account["username"])
    
        password_input = driver.find_element(By.NAME, "ion-input-1")
        password_input.clear()
        password_input.send_keys(account["password"])
        
        login_button = driver.find_element(By.XPATH, "//ion-button[@type='submit']")
        login_button.click()
        
        time.sleep(25)
    
    except Exception as e:
        print(f"Error occurred for account {account['username']}: {e}")

driver.quit()
```

### 30min Cost

```python
file_path = 'accounts.txt'

accounts = []
with open(file_path, 'r') as file:
    lines = file.readlines()
    for i in range(1, len(lines)):  
        splitList = lines[i].strip().split(",") 
        if len(splitList) == 2: 
            accounts.append({"username": splitList[0], "password": splitList[1]})

options = Options()
options.add_argument("--start-maximized") 
service = Service("C:/Users/User/Documents/chromedriver-win64/chromedriver-win64/chromedriver.exe") 

driver = webdriver.Chrome(service=service, options=options)
login_url = "https://bright.glowmarkt.com/login"

for account in accounts:
    try:
        driver.get(login_url)
        WebDriverWait(driver, 7).until(
            EC.visibility_of_element_located((By.NAME, "ion-input-0"))
        )
        
        username_input = driver.find_element(By.NAME, "ion-input-0")
        username_input.clear()
        username_input.send_keys(account["username"])
    
        password_input = driver.find_element(By.NAME, "ion-input-1")
        password_input.clear()
        password_input.send_keys(account["password"])
        
        login_button = driver.find_element(By.XPATH, "//ion-button[@type='submit']")
        login_button.click()
        
        time.sleep(5)
        
        electricity_button = WebDriverWait(driver, 7).until(
            EC.element_to_be_clickable((By.XPATH, "//ion-segment-button[@value='electricity']"))
        )
        electricity_button.click()
        
        time.sleep(2) 
        
        time_period_button = WebDriverWait(driver, 7).until(
            EC.element_to_be_clickable((By.XPATH, "//ion-segment-button[contains(text(), '30min')]"))
        )
        time_period_button.click()
        
        time.sleep(5) 
        
        try:
            rows = WebDriverWait(driver, 10).until(
                EC.presence_of_all_elements_located((By.XPATH, "//ion-row[contains(@class, 'odd-row-style-class') or contains(@class, 'even-row-style-class')]"))
            )
            table_data = []

            for row in rows:
                date_cell = row.find_element(By.XPATH, ".//ion-col[contains(@class, 'historic-list-date-column-values')]")
                cost_cell = row.find_element(By.XPATH, ".//ion-col[contains(@class, 'historic-list-values')]")
                table_data.append({
                    "Date": date_cell.text.strip(),
                    "Cost": cost_cell.text.strip().replace("\u00a3", "") 
                })

            full_data = {
                "name": account["username"],
                "from": table_data[0]["Date"] if table_data else None,
                "to": table_data[-1]["Date"] if table_data else None,
                "period": "30min",
                "data": table_data
            }
            
            output_file = f"{account['username']}_table_data.json"
            with open(output_file, 'w') as json_file:
                json.dump(full_data, json_file, indent=4)
            
            print(f"Data for {account['username']} saved to {output_file}")
        
        except Exception as table_error:
            print(f"Error for {account['username']}: {table_error}")
    
    except Exception as e:
        print(f"Error for account {account['username']}: {e}")

driver.quit()
```

### 30min Energy [kWh] 

```python
file_path = 'accounts.txt'

accounts = []
with open(file_path, 'r') as file:
    lines = file.readlines()
    for i in range(1, len(lines)):  
        splitList = lines[i].strip().split(",") 
        if len(splitList) == 2: 
            accounts.append({"username": splitList[0], "password": splitList[1]})

options = Options()
options.add_argument("--start-maximized") 
service = Service("C:/Users/User/Documents/chromedriver-win64/chromedriver-win64/chromedriver.exe") 

driver = webdriver.Chrome(service=service, options=options)
login_url = "https://bright.glowmarkt.com/login"
name_type = "Energy Consumption(KWh)"
time_period = "30min"

for account in accounts:
    try:
        driver.get(login_url)
        WebDriverWait(driver, 10).until(
            EC.visibility_of_element_located((By.NAME, "ion-input-0"))
        )
        
        username_input = driver.find_element(By.NAME, "ion-input-0")
        username_input.clear()
        username_input.send_keys(account["username"])
    
        password_input = driver.find_element(By.NAME, "ion-input-1")
        password_input.clear()
        password_input.send_keys(account["password"])
        
        login_button = driver.find_element(By.XPATH, "//ion-button[@type='submit']")
        login_button.click()
        
        time.sleep(9)
        
        electricity_button = WebDriverWait(driver, 7).until(
            EC.element_to_be_clickable((By.XPATH, "//ion-segment-button[@value='electricity']"))
        )
        electricity_button.click()
        
        time.sleep(5) 
        wait_time = 15

        ellipsis_button = WebDriverWait(driver, wait_time).until(
            EC.element_to_be_clickable((By.XPATH, "//ion-button[@slot='icon-only']"))
        )
        ellipsis_button.click()

        popover = WebDriverWait(driver, 15).until(
            EC.visibility_of_element_located((By.XPATH, "//ion-list[contains(@class, 'ion-list-info-popover')]"))
        )

        energy_option = WebDriverWait(driver, 30).until(
            EC.element_to_be_clickable((By.XPATH, "//ion-item[ion-label[text()='Energy']]"))
        )
        energy_option.click()

        button_count = driver.execute_script("""
            return document.querySelectorAll('ion-segment-button').length;
        """)
        print(button_count)
        
        buttons_text = driver.execute_script("""
            let buttons = document.querySelectorAll('ion-segment-button');
            let buttonTexts = [];
            buttons.forEach(button => {
                buttonTexts.push(button.textContent.trim());
            });
            return buttonTexts;
        """)
        print("All buttons text:")
        for i, button in enumerate(buttons_text):
            print(f"Button {i+1}: {button}")

        if button_count == 9:
            js_path = """
                document.querySelector('ion-segment-button:nth-of-type(2)').click();
            """
        elif button_count == 8:
            js_path = """
                document.querySelector('ion-segment-button:nth-of-type(1)').click();
            """
        else:
            print("Unexpected number of buttons, expected 8 or 9.")

        driver.execute_script(js_path)

        try:
            rows = WebDriverWait(driver, 10).until(
                EC.presence_of_all_elements_located((By.XPATH, "//ion-row[contains(@class, 'odd-row-style-class') or contains(@class, 'even-row-style-class')]"))
            )
            table_data = []

            for row in rows:
                date_cell = row.find_element(By.XPATH, ".//ion-col[contains(@class, 'historic-list-date-column-values')]")
                cost_cell = row.find_element(By.XPATH, ".//ion-col[contains(@class, 'historic-list-values')]")
                table_data.append({
                    "Date": date_cell.text.strip(),
                    name_type: cost_cell.text.strip().replace("\u00a3", "")
                })
                
            if not table_data:
                full_data = {
                    "name": account["username"],
                    "from": None,
                    "to": None,
                    "period": time_period,
                    "data": [],
                    "message": "No data available!"
                }
            else:
                full_data = {
                    "name": account["username"],
                    "from": table_data[0]["Date"],
                    "to": table_data[-1]["Date"],
                    "period": time_period,
                    "data": table_data
                }

        except Exception as table_error:
            full_data = {
                "name": account["username"],
                "from": None,
                "to": None,
                "period": time_period,
                "data": [],
                "message": "No data available!"
            }

        output_file = f"{account['username']}_{name_type}_data.json"
        with open(output_file, 'w') as json_file:
            json.dump(full_data, json_file, indent=4)

        print(f"Data for {account['username']} saved to {output_file}")

    except Exception as e:
        print(f"Error for account {account['username']}: {e}")

driver.quit()
```
